```python
import pandas as pd
import numpy as np

# Define your input and output paths for each file
ops_data_input_path = #your path
ops_data_output_path = #your path

products_data_input_path = #your path
products_data_output_path = #your path

sales_data_input_path = #your path
sales_data_output_path = #your path

shipping_data_input_path = #your path
shipping_data_output_path = #your path

# Load the operational data file
df_ops = pd.read_csv(ops_data_input_path)

# Make the machine_id into a string
df_ops['machine_id'] = df_ops['machine_id'].astype(str)

# Generate unique two-letter code for each machine_id
machine_codes = {str(i): chr(65 + i // 26) + chr(65 + i % 26) for i in range(1, 27)}

# Create a new column 'machine_code' by mapping machine_id to machine_codes
df_ops['machine_code'] = df_ops['machine_id'].map(machine_codes)

# Save the dirty operational data file
df_ops.to_csv(ops_data_output_path, index=False)


# Load the products data file
df_products = pd.read_csv(products_data_input_path)

# Add "#NK" prefix to product_id
df_products['product_id'] = "#NK" + df_products['product_id'].astype(str)

# Add "-" suffix to about 1/8 of the product_id's
mask = np.random.rand(len(df_products)) < 1/8
df_products.loc[mask, 'product_id'] = df_products.loc[mask, 'product_id'].astype(str) + "-"

# Remove the 's' from the end of 'category' for between 22% and 28% of the records
mask = (np.random.rand(len(df_products)) < 0.28) & (df_products['category'].str.endswith('s'))
df_products.loc[mask, 'category'] = df_products.loc[mask, 'category'].str[:-1]

# Add '-' before 'subcategory' for rows that have 'Accessories'
df_products.loc[df_products['subcategory'] == 'Accessories', 'subcategory'] = '-' + df_products['subcategory']

# Add '$' before half of the 'price' values
mask = np.random.rand(len(df_products)) < 0.5
df_products.loc[mask, 'price'] = "$" + df_products.loc[mask, 'price'].astype(str)

# Save the dirty products data file
df_products.to_csv(products_data_output_path, index=False)


# Load the sales data file
df_sales = pd.read_csv(sales_data_input_path)

# Make 48% of the 'store' entries lower case
mask = np.random.rand(len(df_sales)) < 0.48
df_sales.loc[mask, 'store'] = df_sales.loc[mask, 'store'].str.lower()

# Remove 1/3 of the 'quantity' values
mask = np.random.rand(len(df_sales)) < 1/3
df_sales.loc[mask, 'quantity'] = np.nan

# Save the dirty sales data file
df_sales.to_csv(sales_data_output_path, index=False)


# Load the shipping data file
df_shipping = pd.read_csv(shipping_data_input_path)

# Replace 55% of the True values in the 'error' column with "error"
mask = (np.random.rand(len(df_shipping)) < 0.55) & (df_shipping['error'] == True)
df_shipping.loc[mask, 'error'] = 'error'

# Replace 35% of the False values in the 'error' column with nulls
mask = (np.random.rand(len(df_shipping)) < 0.35) & (df_shipping['error'] == False)
df_shipping.loc[mask, 'error'] = np.nan

# Reformat 'shipping_end' column values to put a "." between each number
df_shipping['shipping_end'] = pd.to_datetime(df_shipping['shipping_end']).dt.strftime('%m.%d.%Y.%H.%M')

# Add a unique 'shipping_id' column to the data
df_shipping['shipping_id'] = range(1, len(df_shipping) + 1)

# Save the dirty shipping data file
df_shipping.to_csv(shipping_data_output_path, index=False)

```
